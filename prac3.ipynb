```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
display(df.info())
display(df.describe())
display(df.shape)
df.columns
```

```python
print('mean:\n',df.mean(numeric_only=True))
print('median:\n',df.median(numeric_only=True))
print('min:\n',df.min(numeric_only=True))
print('max:\n',df.max(numeric_only=True))
print('std:\n',df.std(numeric_only=True))
print('var:\n',df.var(numeric_only=True))
```

```python
grouped = df.groupby('Species')
```

```python
print('Mean',grouped['SepalLengthCm'].mean())
```

```python
print('mean:\n',grouped.mean(numeric_only=True))
print('\nmedian:\n',grouped.median(numeric_only=True))
print('\nmin:\n',grouped.min(numeric_only=True))
print('\nmax:\n',grouped.max(numeric_only=True))
print('\nstd:\n',grouped.std(numeric_only=True))
print('\nvar:\n',grouped.var(numeric_only=True))
```
