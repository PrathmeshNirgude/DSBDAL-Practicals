```python
import pandas as pd
import numpy as np
```

```python
data={
    "Name":["A","B","c","D","E"],
    "Marks":[40,85,np.nan,95,1000],
    "Age":[22,35,28,np.nan,26],
    "Gender":["Male","female","m","f","male"]
}

df = pd.DataFrame(data)
print("dataset:")
print(df)
```

```python
df.isnull().sum()
```

```python
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df['Age'] =df['Age'].fillna(df['Age'].mean())
```

```python
df['Gender'] = df['Gender'].str.lower().replace({'f':'female','m':'male'})
```

```python
print(df)
```

```python
Q3 = df['Marks'].quantile(0.75)
Q1 = df['Marks'].quantile(0.25)
```

```python
IQR = Q3 - Q1
```

```python
lower = Q1-1.5*IQR
upper = Q3+1.5*IQR
print(lower,upper)
```

```python
outlier = df[(df['Marks']<lower) | (df['Marks']>upper)]
print(outlier)
```

```python
df = df[(df['Marks']>lower)&(df['Marks']<upper)]
```

```python
print(df)
```

```python
df['marks_log'] = np.log(df['Marks'])
```

```python
df['Marks_norm'] = (df['Marks']-df['Marks'].min())/(df['Marks'].max()-df['Marks'].min())
```

```python
df['Marks_scaling'] = (df['Marks'] - df['Marks'].mean())/df['Marks'].std()
```

```python
print(df)
```

```python
# ----------------------------
# 4. Z-score calculation (outlier detection)
# ----------------------------
df['z_score'] = (df['Marks'] - df['Marks'].mean()) / df['Marks'].std()

print("\nWith Z-score:")
print(df)

# ----------------------------
# 5. Remove outliers (|Z| > 3)
# ----------------------------
df = df[df['z_score'].abs() <= 3]

print("\nAfter removing outliers:")
print(df)
```
